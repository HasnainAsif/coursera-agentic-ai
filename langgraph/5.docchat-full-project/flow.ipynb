**What is docling:**

- Its more flexible than PdfReader of LangChain. It can read PNG with Images due to builtIn OCR and PDFs saved as images.
- Its not just specified on PDFs, it can also read from docs, txt, md etc

### document_processor/file_handler.py
- Its used to extract content from pdf documents. Convert that content to markdown and then create chunks.

### retriever/builder.py
- Its used to create hybrid retriever. BM25(exact-keyword-matching) + vector search retriever(For semantic search)

### agents/workflow.py
- Complete LangGraph workflow, like nodes, edges etc
