# Simple Chat Messages and comparing different models

```python
# Suppress warnings (same as before)
def warn(*args, **kwargs):
    pass

import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')

import os
os.environ['ANONYMIZED_TELEMETRY'] = 'False'

# ---- OpenAI + LangChain imports ----
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage # for creating message objects with specific roles
from langchain_core.output_parsers import StrOutputParser
```

```python
# Model configuration
openai_llm = ChatOpenAI(
    model="gpt-4.1-nano",
    temperature=0.2,
    max_tokens=256
)
```

```python
response = openai_llm.invoke([
    HumanMessage(content="What is 2+2?")
])
print(response.content)
```

## Chat message

```python
msg = openai_llm.invoke(
    [
        SystemMessage(content="You are a supportive AI bot that suggests fitness activities to a user in one short sentence"), # System prompt
        HumanMessage(content="I like high-intensity workouts, what should I do?"), # User query
        AIMessage(content="You should try a CrossFit class"), # AI reply
        HumanMessage(content="How often should I attend?") # Follow-up user query
    ]
)
```

### Compare Model Responses with Different Parameters

- Compare responses of different models
- Change temperature and check how temperature affects responses

Check out these prompt types:

| Prompt type |   Prompt Example  |
|------------------- |--------------------------|
| **Creative writing**  | "Write a short poem about artificial intelligence." |
| **Factual questions** |  "What are the key components of a neural network?"  |
| **Instruction-following**  | "List 5 tips for effective time management." |

```python
creative_params = {
    "temperature": 0.8, # Higher temperature for more creative responses
    "max_tokens": 256,
}

precise_params = {
    "temperature": 0.1, # # Lower temperature for more deterministic(precise) responses
    "max_tokens": 256,
}

model_1_creative = ChatOpenAI(
    model="gpt-4o-mini",
    **creative_params
)
model_1_precise = ChatOpenAI(
    model="gpt-4o-mini",
    **precise_params
)

model_2_creative = ChatOpenAI(
    model="gpt-4.1-nano",
    **creative_params
)
model_2_precise = ChatOpenAI(
    model="gpt-4.1-nano",
    **precise_params
)

prompt = "Explain AI Agents in simple terms." # Somehow related to Factual question

response_1_creative = model_1_creative.invoke([HumanMessage(content=prompt)])
response_1_precise = model_1_precise.invoke([HumanMessage(content=prompt)])
response_2_creative = model_2_creative.invoke([HumanMessage(content=prompt)])
response_2_precise = model_2_precise.invoke([HumanMessage(content=prompt)])

print("GPT-4o-mini (creative):\n", response_1_creative.content)
print("\nGPT-4o-mini (precise):\n", response_1_precise.content)
print("\nGPT-4.1-nano (creative):\n", response_2_creative.content)
print("\nGPT-4.1-nano (precise):\n", response_2_precise.content)
```

According to above responses: I am declined towards `High Temperature` responses and for models: gpt-4o-mini
